```python
import matplotlib.pyplot as plt

# Datos
años = [1961, 1965, 1969, 1973, 1989, 1993, 1997, 2001, 2005, 2009, 2013, 2017, 2021]
porcentaje_votos = [14.07, 46.36, 32.71, 36.18, 32.18, 20.22, 29.22, 22.84, 29.72, 16.63, 16.50, 14.29, 4.59]


# Gráfico
plt.plot(años, porcentaje_votos, marker='o', color='pink', label='Porcentaje de Votos')
plt.xlabel('Año de Elección')
plt.ylabel('Porcentaje de Votos Obtenidos')
plt.title('Porcentaje de votos obtenidos por senadores DC por año')
plt.xticks(años, rotation=45)
plt.grid(True)

# Leyendas
plt.legend(loc='upper right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
