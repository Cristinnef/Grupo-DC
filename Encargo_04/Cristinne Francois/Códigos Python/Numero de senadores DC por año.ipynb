```python
import matplotlib.pyplot as plt

# Datos
años = [1961, 1965, 1969, 1973, 1989, 1993, 1997, 2001, 2005, 2009, 2013, 2017, 2021]
num_senadores = [2, 10, 9, 10, 12, 4, 9, 2, 5, 4, 2, 3, 2]

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Gráfico
plt.bar(años, num_senadores, color='pink')
plt.xlabel('Año de Elección')
plt.ylabel('Número de Senadores')
plt.title('Número de senadores DC electos por año')
plt.xticks(años, rotation=45)


# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
