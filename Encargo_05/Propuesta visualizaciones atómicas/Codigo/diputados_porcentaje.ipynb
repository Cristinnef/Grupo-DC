```python
import matplotlib.pyplot as plt

# Datos
aos = [1961, 1965, 1969, 1973, 1989, 1993, 1997, 2001, 2005, 2009, 2013, 2017, 2021]
porcentaje_votos = [15.40, 42.30, 29.80, 28.70, 25.99, 27.12, 22.98, 18.92, 20.76, 14.21, 15.55, 10.28, 4.17]

# Grfico
plt.plot(aos, porcentaje_votos, marker='o', color='orange', label='Porcentaje de Votos')
plt.xlabel('Ao de Eleccin')
plt.ylabel('Porcentaje de Votos Obtenidos')
plt.title('Porcentaje de votos obtenidos por diputados DC por ao')
plt.xticks(aos, rotation=45)
plt.grid(True)

# Leyendas
plt.legend(loc='upper right')

# Mostrar el grfico
plt.tight_layout()
plt.show()
```
