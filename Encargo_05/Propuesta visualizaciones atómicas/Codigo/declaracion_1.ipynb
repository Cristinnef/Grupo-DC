```python
# Importar Matplot
import matplotlib.pyplot as plt
import numpy as np

#Definir variables y valores

Años = ("2017", "2018", "2019", "2022", "2023")
Políticos = (1, 4, 1, 5, 3)

# Gráfico de barras
fig, ax = plt.subplots()
ax.bar(x = Años, height = Políticos, color = "#f8766d", align = "edge", width = 0.4)

# Definir las etiquetas y el título del gráfico
ax.set_ylabel('Políticos')
ax.set_xlabel('Años')
ax.set_title('Miembros de la Democracia Cristiana que ingresaron su declaración de asunción al cargo')

# plt.show()
```
