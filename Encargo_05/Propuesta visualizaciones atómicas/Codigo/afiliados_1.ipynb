```python
# Importar matplot
import matplotlib.pyplot as plt

# Crear nueva figura y ejes
fig, ax = plt.subplots()

# Definir valores
partidos = ['PC', 'PDG', 'PS', 'RN', 'CS', 'UDI', 'DC', 'PPD', 'RD', 'PR']
afiliados = [45.882, 42.128, 40.312, 39.395, 36.767, 34.121, 29.099, 27.826, 26.624, 26.353]

# Definir colores
bar_labels = ['red', 'red', 'red', 'red', 'red', 'red', 'blue', 'red', 'red', 'red' ]
bar_colors = ['tab:red', 'tab:red', 'tab:red', 'tab:red', 'tab:red', 'tab:red', 'tab:blue', 'tab:red', 'tab:red', 'tab:red']

# Generar gráfico de barras
ax.bar(partidos, afiliados, label=bar_labels, color=bar_colors)

# Agregar título y etiqueta eje 'Y'
ax.set_ylabel('Miles de afiliados')
ax.set_title('10 partidos políticos con más afiliados en 2023')

# Mostrar el gráfico
plt.show()
```
