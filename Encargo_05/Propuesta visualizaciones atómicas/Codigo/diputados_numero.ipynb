```python
import matplotlib.pyplot as plt

# Datos
años = [1961, 1965, 1969, 1973, 1989, 1993, 1997, 2001, 2005, 2009, 2013, 2017, 2021]
num_diputados = [23, 82, 56, 50, 38, 37, 38, 23, 20, 19, 21, 14, 8]

# Crear la figura y los ejes
plt.figure(figsize=(10, 5))

# Configurar los ejes
plt.bar(años, num_diputados, color='orange')
plt.xlabel('Año de Elección')
plt.ylabel('Número de Diputados')
plt.title('Número de diputados DC electos por año')
plt.xticks(años, rotation=45)


# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
